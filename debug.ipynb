### Check data statistics

```python
from pathlib import Path
import numpy as np
import json
```

```python
dataset_name = "ADT"
data_dir = Path(f"/home/mahatano/workspace/mdm/dataset/{dataset_name}/new_joints")
prime_dir = Path(f"/home/mahatano/workspace/mdm/dataset/{dataset_name}/priming/output/3d_primed_locations")

color_map = {
    "ADT": "blue",
    "ParaHome": "orange",
    "MoGaze": "purple"
}

root_dist_list = []
hand_dist_list = [] # right wrist
prime_gap_list = []

for data_path in data_dir.glob("*.npy"):
    data = np.load(data_path)
    
    # calculate maximum distance of root joint in each sequence
    root_positions = data[:, 0]
    rhand_positions = data[:, 21] - root_positions  # right wrist relative to root
    max_distance_root = np.max(np.linalg.norm(root_positions - root_positions[0], axis=1))
    max_distance_hand = np.max(np.linalg.norm(rhand_positions - rhand_positions[0], axis=1))
    root_dist_list.append(max_distance_root)
    hand_dist_list.append(max_distance_hand)

if dataset_name != "ParaHome":
    for prime_path in prime_dir.glob("*.json"):
        with open(prime_path) as f:
            prime_data = json.load(f)
        video_id = prime_path.stem
        for vid, prime_dict in prime_data[video_id].items():
            if prime_dict["start"] != "":
                frame_primed = prime_dict["start"]["prime_stats"]["frame_primed"]
                if frame_primed >= 0:
                    prime_gap_list.append(prime_dict["start"]["prime_stats"]["prime_gap"])
            if prime_dict["end"] != "":
                frame_primed = prime_dict["end"]["prime_stats"]["frame_primed"]
                if frame_primed >= 0:
                    prime_gap_list.append(prime_dict["end"]["prime_stats"]["prime_gap"])

print("="*20)
print("Root distance")
print("Max:", np.max(root_dist_list))
print("Min:", np.min(root_dist_list))
print("Mean:", np.mean(root_dist_list))
print("Std:", np.std(root_dist_list))

print("="*20)
print("Hand distance")
print("Max:", np.max(hand_dist_list))
print("Min:", np.min(hand_dist_list))
print("Mean:", np.mean(hand_dist_list))
print("Std:", np.std(hand_dist_list))

if dataset_name != "ParaHome":
    print("="*20)
    print("Prime gap")
    print("Max:", np.max(prime_gap_list))
    print("Min:", np.min(prime_gap_list))
    print("Mean:", np.mean(prime_gap_list))
    print("Std:", np.std(prime_gap_list))

print(len(root_dist_list), len(hand_dist_list))

# create histogram plot for each: change color, every 0.1m interval
import matplotlib.pyplot as plt
# plt.hist(root_dist_list, bins=np.arange(0, np.max(root_dist_list) + 0.05, 0.05), alpha=0.5, label='Root Joint', color=color_map[dataset_name])
plt.hist(root_dist_list, bins=100, alpha=0.5, label='Pelvis Joint', color=color_map[dataset_name])
plt.xlabel('Displacement (m)')
plt.legend()
plt.savefig(f"root_histogram_{dataset_name}.png")
plt.clf()

# plt.hist(hand_dist_list, bins=np.arange(0, np.max(hand_dist_list) + 0.05, 0.05), alpha=0.5, label='Right Hand (Wrist)', color=color_map[dataset_name])
plt.hist(hand_dist_list, bins=100, alpha=0.5, label='Hand (Wrist)', color=color_map[dataset_name])
plt.xlabel('Displacement (m)')
plt.legend()
plt.savefig(f"rhand_histogram_{dataset_name}.png")
plt.clf()

if dataset_name != "ParaHome":
    plt.hist(prime_gap_list, bins=100, alpha=0.5, label='Prime Gap', color=color_map[dataset_name])
    plt.xlabel('Prime gap (s)')
    plt.legend()
    plt.savefig(f"prime_gap_histogram_{dataset_name}.png")
    plt.clf()
```

```python
from pathlib import Path
import numpy as np
from tqdm import tqdm
import json

dataset_name = "MoGaze"

new_joints_vec_dir = Path(f"./dataset/{dataset_name}/new_joint_vecs")
data_dir = Path(f"./dataset/{dataset_name}/priming/output/3d_primed_locations")
save_dir = Path(f"./dataset/{dataset_name}/texts")
save_dir.mkdir(exist_ok=True, parents=True)

for new_joint_vecs_path in tqdm(new_joints_vec_dir.glob("*.npy")):
    video_id = "_".join(new_joint_vecs_path.name.split("_")[:-2])
    seq_start = int(new_joint_vecs_path.stem.split("_")[-2])
    seq_end = int(new_joint_vecs_path.stem.split("_")[-1])
    json_path = data_dir / f"{video_id}.json"
    
    if not json_path.exists():
        print("No json:", video_id)
    else:
        with open(json_path) as f:
            data = json.load(f)
        for id, prime_dict in data[video_id].items():
            timesegment_start = int(prime_dict["timesegment"][0])
            timesegment_end = int(prime_dict["timesegment"][1])
            type = None
            if timesegment_start == seq_end:
                type = "pick"
            elif timesegment_end == seq_end:
                type = "put"
            if type is not None:
                text = f"C moves to {type} object"
                save_path = save_dir / f"{video_id}_{seq_start}_{seq_end}.txt"
                save_path_m = save_dir / f"M{video_id}_{seq_start}_{seq_end}.txt"
                with open(save_path, "w") as f:
                    f.write(text + "\n")
                with open(save_path_m, "w") as f:
                    f.write(text + "\n")
```
